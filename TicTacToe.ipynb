```python
emptyBoard = (
'''
     |     |     
     |     |    
_____|_____|_____
     |     |     
     |     |     
_____|_____|_____
     |     |     
     |     |   
     |     |     
'''
    )



availableMoves = ['top left', 'top', 'top right', 'left', 'centre', 'right', 'bottom left', 'bottom', 'bottom right']

currentBoardState = emptyBoard
currentBoardStateList = list(currentBoardState)

XplayerMoves = []
OplayerMoves = []


winConditionLeftCol = {'top left', 'left', 'bottom left'}
winConditionMiddleCol = {'top', 'centre', 'bottom'}
winConditionRightCol = {'top right', 'right', 'bottom right'}

winConditionTopRow = {'top left', 'top', 'top right'}
winConditionMiddleRow = {'left', 'centre', 'right'}
winConditionBottomRow = {'bottom left', 'bottom', 'bottom right'}

winConditionDiag = {'top left', 'centre', 'bottom right'}
winConditionOffDiag = {'top right', 'centre', 'bottom left'}

winConditions = [winConditionLeftCol,winConditionMiddleCol,winConditionRightCol,
                winConditionTopRow, winConditionMiddleRow, winConditionBottomRow,
                winConditionDiag, winConditionOffDiag]

GameWon = False     #A flag for if the game is won or a draw.

while len(availableMoves) > 0:
    
    if len(availableMoves) == 9:
        currentBoardState = emptyBoard
        print ('''Player X goes first. Where would you like to play? The moves are called: 
'top left', 'top', 'top right', 
'left', 'centre', 'right', 
'bottom left', 'bottom' and 'bottom right'.''')
        
    print currentBoardState
    
    
    if len(availableMoves)%2 == 1:
        player = 'Xplayer'
        playerPrompt = "It is player X's turn"
    else:
        player = 'Oplayer'
        playerPrompt = "It is player O's turn"
        
    if len(availableMoves) < 9:
        print playerPrompt
        
    currentMove = raw_input()

    if currentMove == 'top left' and currentMove in availableMoves:
        if player == 'Xplayer':
            currentBoardStateList.insert(21,'X')
            currentBoardStateList.pop(22)
        elif player == 'Oplayer':
            currentBoardStateList.insert(21,'O')
            currentBoardStateList.pop(22)
    elif currentMove == 'top' and currentMove in availableMoves:
        if player == 'Xplayer':
            currentBoardStateList.insert(27,'X')
            currentBoardStateList.pop(28)
        elif player == 'Oplayer':
            currentBoardStateList.insert(27,'O')
            currentBoardStateList.pop(28)
    elif currentMove == 'top right' and currentMove in availableMoves:
        if player == 'Xplayer':
            currentBoardStateList.insert(33,'X')
            currentBoardStateList.pop(34)
        elif player == 'Oplayer':
            currentBoardStateList.insert(33,'O')
            currentBoardStateList.pop(34)
    elif currentMove == 'left' and currentMove in availableMoves:
        if player == 'Xplayer':
            currentBoardStateList.insert(74,'X')
            currentBoardStateList.pop(75)
        elif player == 'Oplayer':
            currentBoardStateList.insert(74,'O')
            currentBoardStateList.pop(75)
    elif currentMove == 'centre' and currentMove in availableMoves:
        if player == 'Xplayer':
            currentBoardStateList.insert(80,'X')
            currentBoardStateList.pop(81)
        elif player == 'Oplayer':
            currentBoardStateList.insert(80,'O')
            currentBoardStateList.pop(81)
    elif currentMove == 'right' and currentMove in availableMoves:
        if player == 'Xplayer':
            currentBoardStateList.insert(86,'X')
            currentBoardStateList.pop(87)
        elif player == 'Oplayer':
            currentBoardStateList.insert(86,'O')
            currentBoardStateList.pop(87)
    elif currentMove == 'bottom left' and currentMove in availableMoves:
        if player == 'Xplayer':
            currentBoardStateList.insert(128,'X')
            currentBoardStateList.pop(129)
        elif player == 'Oplayer':
            currentBoardStateList.insert(128,'O')
            currentBoardStateList.pop(129)
    elif currentMove == 'bottom' and currentMove in availableMoves:
        if player == 'Xplayer':
            currentBoardStateList.insert(134,'X')
            currentBoardStateList.pop(135)
        elif player == 'Oplayer':
            currentBoardStateList.insert(134,'O')
            currentBoardStateList.pop(135)
    elif currentMove == 'bottom right' and currentMove in availableMoves:
        if player == 'Xplayer':
            currentBoardStateList.insert(140,'X')
            currentBoardStateList.pop(141)
        elif player == 'Oplayer':
            currentBoardStateList.insert(140,'O')
            currentBoardStateList.pop(141)
    else:
        print 'The move you entered is not valid. The valid current moves left are ', availableMoves, '.'
            
        continue
    
    if player == 'Xplayer':
        XplayerMoves.append(currentMove)
    elif player == 'Oplayer':
        OplayerMoves.append(currentMove)
        
    availableMoves.remove(currentMove)
    
    for x in range(0,8):
        if winConditions[x] <= set(OplayerMoves):
            print 'Player O wins!'
            availableMoves = []
            GameWon = True
            
    for x in range(0,8):
        if winConditions[x] <= set(XplayerMoves):
            print 'Player X wins!'
            availableMoves = []
            GameWon = True

    currentBoardState = ''.join(currentBoardStateList)

print currentBoardState 

if GameWon == False:
    print "It's a draw!"
```
